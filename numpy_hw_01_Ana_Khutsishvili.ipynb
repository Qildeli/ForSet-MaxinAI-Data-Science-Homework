```python
import numpy as np
import matplotlib.pyplot as plt
```

# 1

სიმულირებული `(x1, y1)` წყვილებისთვის დაამოდელირეთ პოლინომური რეგრესიები და შეადარეთ ცდომილებები.




```python
num_dots = 200

np.random.seed(42)

x1 = np.random.rand(num_dots) * 6 - 3
y1 = np.zeros(num_dots)

for i, p in enumerate([-3, 0, 3, 1]):
    y1 = y1 + np.power(x1, i) * p

y1 = y1 + np.random.normal(loc=0, scale=5, size=num_dots)
plt.scatter(x1, y1)
plt.show()
```

```python
poly_1 = np.polyfit(x1, y1, 1)
poly_2 = np.polyfit(x1, y1, 2)
poly_3 = np.polyfit(x1, y1, 3)

poly_1 = np.poly1d(poly_1)
poly_2 = np.poly1d(poly_2)
poly_3 = np.poly1d(poly_3)

poly_1.c, poly_2.c, poly_3.c

yhat_1 = poly_1(x1)
yhat_2 = poly_2(x1)
yhat_3 = poly_3(x1)
```

```python
plt.scatter(x1, y1)

plt.scatter(x1, yhat_1)

plt.show()
```

```python
plt.scatter(x1, y1)

plt.scatter(x1, yhat_2)

plt.show()
```

```python
plt.scatter(x1, y1)

plt.scatter(x1, yhat_3)

plt.show()
```

```python
tss = np.power(y1 - y1.mean(), 2).sum()

rss1, rss2, rss3 = np.power(y1 - yhat_1, 2).sum(), np.power(y1 - yhat_2, 2).sum(), np.power(y1 - yhat_3, 2).sum()

#R^2

1-rss1/tss, 1-rss2/tss, 1-rss3/tss
```

$$
\begin{cases}
x+2y=11\\
5x-3y=3
\end{cases}

```python
a = [[1, 2],
     [5, 3]]

b = [[11],
     [3]]

z = np.linalg.solve(a,b)

print(z)
```

# 5

მოცემული მასივისთვის ელემენტებს შორის ჩაამატეთ 3-3 ნულები:

`[1, 2, 3]` -> `[1, 0, 0, 0, 2, 0, 0, 0, 3]`

`[1, 2]` -> `[1, 0, 0, 0, 2]`

```python
arr1 = np.array([1, 2, 3])


n = len(arr1)
arr = np.zeros((4*n - 3))
arr[range(0, 4*n, 4)] = arr1

print(arr)
```
